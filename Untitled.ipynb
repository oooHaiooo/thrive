```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer

class AgeImputer(BaseEstimator, TransformerMixin):
      #自定义转换器  BaseEstimator估计器  TransformerMixin 是 Scikit-Learn 中的一个混合类，用于所有转换器。它提供了两个主要功能：一个 fit_transform 方法和一个 set_output 方法
 
    
    def fit(self, X, y=None):
        #  参数:
        #X : DataFrame 或类似数组的结构，包含需要转换的数据。
        #y : 通常用于监督学习，但在这个转换器中不使用。
        #这里通常可以不做任何操作，因为我们使用的是均值填充，不需要拟合数据
        return self
        # 返回:self : 返回转换器的实例，以便可以进行链式调用
        
    def transform(self, X):
        #参数:X : DataFrame 或类似数组的结构，包含需要转换的数据
        #transform 方法用于将输入数据 X 转换为新的数据，在这个例子中，就是将 Age 列中的缺失值用均值填充。
        
        imputer = SimpleImputer(strategy="mean")
        #SimpleImputer 来填充 Age 列中的缺失值，策略为均值填充。
       
        X['Age'] = imputer.fit_transform(X[['Age']])
        return X
```

```python
from sklearn.preprocessing import OneHotEncoder
class FeatureEncoder(BaseEstimator, TransformerMixin):
    #FeatureEncoder的类，它继承自BaseEstimator和TransformerMixin，这两个类都是scikit-learn库中的一部分，用于创建可以在scikit-learn的Pipeline中使用的转换器。
    def fit(self,X,y=None):
        return self

    def transform(self,X):
        encoder =  OneHotEncoder()
        # 创建OneHotEncoder实例会处理输入数据中的分类特征，将它们转换为二进制（0和1）的列
        
        matrix = encoder.fit_transform(X[['Embarked']]).toarray()
        #toarray 方法的作用是将稀疏矩阵转换为一个标准的密集矩阵（dense matrix），即一个常规的二维数组，其中所有元素（包括零）都被显式地存储。
        column_names = ["C","S","Q","N"]#定义别名列表
        for i in range(len(matrix.T)):
            #.T：是数组的转置。如果matrix是一个二维数组（例如矩阵），那么matrix.T就是将原数组的行和列互换得到的新数组。
            #len()函数用于获取对象的长度，对于二维数组（或矩阵）的转置来说，len(matrix.T)将返回转置后数组的行数（即原数组的列数）。
            #range()函数用于生成一个整数序列。这里，它根据len(matrix.T)的结果生成一个从0开始、到len(matrix.T)-1结束的整数序列。
            X[column_names[i]] = matrix.T[i]
            #转换将字符转换成英文

        matrix = encoder.fit_transform(X[['Sex']]).toarray()
        column_names = ["Female","Male"]
        for i in range(len(matrix.T)):
            X[column_names[i]] = matrix.T[i]
         #同上文一样
        return  X
```

```python
class FeatureDropper(BaseEstimator, TransformerMixin):
    #FeatureDropper 类的目的是在数据预处理阶段，从输入的数据集中删除特定的列。
    def fit(self,X,y=None):
        return self

    def transform(self,X):
        return X.drop(["Embarked","Name","Ticket","Cabin","Sex","N"],axis=1,errors="ignore")
        #前面所写的红色字体是指所要删除的列，axis=1 表示操作是在列上进行的，errors="ignore" 参数表示如果指定的列名不存在，则忽略这个错误而不是抛出异常。
```

```python
from sklearn.pipeline import Pipeline
pipeline=Pipeline([("ageimputer",AgeImputer()),
                   ("featureencoder",FeatureEncoder()),
                    ("featuredropper",FeatureDropper())])
 #构建了一个 Pipeline 对象，该对象串联了三个预处理步骤：ageimputer（年龄填充器）、featureencoder（特征编码器）和 featuredropper（特征丢弃器）                  
```

```python
strat_train_set = pipeline.fit_transform(strat_train_set) #得到最终转换结果返回
```
